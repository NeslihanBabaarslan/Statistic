## Descriptive Statistics

 Import **NumPy**, **SciPy**, and **Pandas**

```
import numpy as np
import pandas as pd
from scipy import stats
```

 Randomly generate 1,000 samples from the normal distribution using `np.random.normal()`(mean = 100, standard deviation = 15)

```
samples = np.random.normal(100,15,1000)
samples
```

Compute the **mean**, **median**, and **mode**

```
mean =round(np.mean(samples),3)
median =round(np.median(samples),3)
mode =stats.mode(samples)
print("mean is {}, median is {} and mode is {}.".format(mean, median, mode))
```

Compute the **min**, **max**, **Q1**, **Q3**, and **interquartile range**

```
min = round(np.min(samples),2)
max =round(np.max(samples),2)
q1 =round(np.percentile(samples, 25),2)
q3 =round(np.percentile(samples, 75),2)
iqr =round(stats.iqr(samples),2)
iqr2=round((q3-q1),2)
print("min= ", min)
print("max= ", max)
print("q1= ", q1)
print("q3= ", q3)
print("iqr: {} and iqr2: {} are equal.".format(iqr,iqr2))
```

Compute the **variance** and **standard deviation**

```
variance = round(np.var(samples),3)
std_dev =round(np.std(samples),3)
print("variance= {}, std_dev= {} ".format(variance, std_dev))
```

Compute the **skewness** and **kurtosis**

```
skewness =round(stats.skew(samples),3)
kurtosis =round(stats.kurtosis(samples),3)
print("skewness= {}, kurtosis= {}".format(skewness, kurtosis))
```

## NumPy Correlation Calculation

Create an array x of integers between 10 (inclusive) and 20 (exclusive). Use `np.arange()`

```
x = np.arange(10,20)
x
```

Then use `np.array()` to create a second array y containing 10 arbitrary integers.

```

y=np.array([13,6,7,34,47,23,56,34,5,1])
y
```

Once you have two arrays of the same length, you can compute the **correlation coefficient** between x and y

```
r = np.corrcoef(x,y)
r
```
